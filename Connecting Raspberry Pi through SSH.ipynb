# Connecting to raspberry pi through ssh 
### Software required :
    -> putty (Windows OS)



Step 1 : <br>Start the terminal on raspberry pi.

Step 2 : <br>Type the following command into the terminal :<br> _ip addr_

Step 3 : <br>Locate the outcome with _wlan0_ prefix. Locate the line of outcome with _inet_, the ip address of the rapsberry pi is the number followed.

As depicted above, the ip address is :
192.168.0.156

Step 4 : <br>
Open up Putty.exe program. Key in the ip address password and click on _Open_ button.


Step 5 : <br>
Afterward, a new terminal will show up.
<br>Key in the following account detail :
<br>user : _pi_
<br>password : _raspberry_


Step 6:
<br>Testing status by type in the following command : 
<br>_whoami_
